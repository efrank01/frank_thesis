```python
%pylab inline
from astropy.io import fits
from astropy.io import ascii
from scipy.io import readsav
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Column, Table, vstack
import tarfile
import os
import glob
from pathlib import Path
from scipy import interpolate
from scipy.interpolate import InterpolatedUnivariateSpline
import pdb
```

```python
drpallfile = '/Users/Frank/KNAC_Internship/drpall-v2_7_1.fits'
outdir = '/Users/Frank/KNAC_Internship/outdir/'

hdu = fits.open(drpallfile)
db = hdu[1].data

#collect GBT spectra
searchpath = os.path.join('/Users/Frank/KNAC_Internship/ALL_samples_spectra','*.fits')
searchnames = glob.glob(searchpath)
#print(searchpath)
for file in glob.glob(searchpath):
    

    #read the file
    hdu = fits.open(file)
    data = hdu[1].data
    hdr = hdu[1].header
    vhi = data['vhi'].squeeze()
    fhi = data['fhi'].squeeze()*1000

    print(hdr['object'])
    vopt = hdr['obj_vel']
    
    #read in the parameter file
    parfile = '/Users/Frank/KNAC_Internship/ALL_samples_spectra/'+hdr['object']+'_par.sav'
    if os.path.exists(parfile):
        par = readsav(parfile)
        regions = par.par[0]['blregions']
    else:
        regions = [[-1,-1]]
    
    fig = plt.figure()

    plt.clf()
    plt.plot(vhi,fhi)
    plt.xlim(vopt-1000,vopt+1000)
    plt.ylim()
    plt.xlabel('velocity [km/s]')
    plt.ylabel('flux density [mJy]')
    plt.title(hdr['object'])
    plt.plot([vopt-1000,vopt+1000],[0,0],':',color='red')
    axes = plt.gca()
    sel=(vhi > (vopt-1000)) & (vhi < (vopt+1000))
    ymin,ymax = np.min(fhi[sel]),np.max(fhi[sel])
    dy = ymax - ymin
    #axes.set_ylim(ymin-0.1*dy,ymax+0.1*dy)
    
    ylim = axes.get_ylim()
    plt.plot([vopt,vopt],[ylim[0],ylim[1]],'--',color='red')
    axes.set_ylim(ylim)
        
    for p in regions:
        if p[0] != -1:
            p=p-([int(150./4),int(150./4)])
            p[np.where(p < 0)] = 0
            p[np.where(p > len(vhi)-1)] = len(vhi)-1
            print(p)
            plt.plot([vhi[p[0]],vhi[p[1]]],[0,0],color='orange',linewidth=5,alpha=0.7)

    plt.savefig(outdir + hdr['object']+'spec4check.png',dpi=300)
            
    plt.close()
```
